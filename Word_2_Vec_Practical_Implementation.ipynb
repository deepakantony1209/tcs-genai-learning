## Understanding Word2Vec using Gensim

This notebook demonstrates how to use the `gensim` library to work with Word2Vec models. Word2Vec is a popular technique for learning word embeddings, which are dense vector representations of words that capture semantic relationships.

```python
! pip install gensim
```

### Install Gensim

First, we need to install the `gensim` library if it's not already installed. This command uses `pip` to install the package.

```python
import gensim
import gensim.downloader as api
from gensim.models import Word2Vec, KeyedVectors
```

### Import Libraries

Next, we import the necessary modules from `gensim`:
- `gensim`: The main library.
- `gensim.downloader as api`: Used to download pre-trained word embedding models.
- `Word2Vec`, `KeyedVectors`: Classes for working with Word2Vec models and their vectors.

```python
wv = api.load('word2vec-google-news-300')
```

### Load a Pre-trained Word2Vec Model

We load a large pre-trained Word2Vec model from `gensim.downloader`. The `'word2vec-google-news-300'` model is trained on a Google News dataset and contains 300-dimensional vectors for approximately 3 million words and phrases.

```python
vec_deepak  = wv['deepak']
```

### Get the Vector for a Specific Word

Once the model is loaded, we can access the vector representation for any word present in the model's vocabulary. Here, we retrieve the vector for the word 'deepak'.

```python
vec_deepak
```

### Display the Word Vector

This cell simply displays the vector for 'deepak'. Word vectors are typically numpy arrays of floating-point numbers.

```python
wv.most_similar('deepak')
```

### Find Most Similar Words

The `most_similar()` method allows us to find words that are semantically similar to a given word based on their vector representations. Words with vectors closer in space are considered more similar.

```python
wv.most_similar('man')
```

### Find Most Similar Words for 'man'

Another example of finding semantically similar words, this time for 'man'.

```python
wv.similarity('deepak', 'lovely')
```

### Calculate Similarity Between Two Words

The `similarity()` method computes the cosine similarity between the vector representations of two words. A higher value indicates greater similarity (closer to 1.0).

```python
wv.similarity('android', 'ios')
```

### Calculate Similarity Between 'android' and 'ios'

Demonstrates the semantic similarity between 'android' and 'ios', which are related concepts in the technology domain.

```python
wv.similarity('sachin', 'dhoni')
```

### Calculate Similarity Between 'sachin' and 'dhoni'

Shows the semantic similarity between two famous cricketers, 'sachin' and 'dhoni', demonstrating how the model captures relationships between proper nouns in the same domain.

```python
wv.most_similar('happy')
```

### Find Most Similar Words for 'happy'

Yet another example of finding similar words, showcasing words with similar emotional connotations to 'happy'.

```python
vec = wv['king']-wv['queen']+wv['woman']
```

### Vector Arithmetic (King - Queen + Woman)

Word2Vec models allow for interesting vector arithmetic. This operation, `king - queen + woman`, is famous for often resulting in a vector that is very similar to 'man', demonstrating how the model learns analogies and relationships between words.

```python
wv.most_similar(vec)
```

### Find Most Similar Word for the Resultant Vector

Finally, we find the words most similar to the `vec` (the result of `king - queen + woman`) to see if the analogy holds. As expected, 'man' often appears among the most similar words.
