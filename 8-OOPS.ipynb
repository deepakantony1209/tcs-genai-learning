Define behavior of objects.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof!")

dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 4)

dog1.bark()
dog2.bark()
```

## 5. Bank Account Example
A real-world OOP implementation.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. New balance: {self.balance}")

    def get_balance(self):
        return self.balance

# Using the class
account = BankAccount("Deepak", 5000)

account.deposit(100)
account.withdraw(300)

print("Final Balance:", account.get_balance())
```

## Inheritance


### Single level inheritance

Single level inheritance is a fundamental concept in object-oriented programming where a new class (called a child or subclass) inherits properties and behaviors (attributes and methods) from an existing class (called a parent or superclass). This allows for code reusability and the creation of more specific classes based on general ones.

Explaining single level inheritance with an example of a Tesla class that inherits from a Car class. The Tesla class will have additional attributes and methods specific to electric cars, while still retaining the properties and behaviors of a general Car.

```python
class Car:
    def __init__(self, model, doors, windows, max_speed):
        self.model = model
        self.doors = doors
        self.windows = windows
        self.max_speed = max_speed
    
    def get_info(self):
        return f"Model: {self.model}, Doors: {self.doors}, Windows: {self.windows}, Max Speed: {self.max_speed} km/h"
    
car1 = Car("Toyota Camry", 4, 4, 240)
print(car1.get_info())
```

```python
class Tesla(Car):
    def __init__(self, model, doors, windows, max_speed, battery_capacity, is_autopilot):
        super().__init__(model, doors, windows, max_speed)
        self.battery_capacity = battery_capacity
        self.is_autopilot = is_autopilot
    
    def get_info(self):
        base_info = super().get_info()
        return f"{base_info}, Battery Capacity: {self.battery_capacity} kWh, Autopilot: {'Yes' if self.is_autopilot else 'No'}"
```

```python
tesla = Tesla("Model S", 4, 4, 250, 100, True)              
```

```python
tesla.get_info()
```

## Multiple Inheritance



```python
## Generate a simple code to exhibit multiple inheritance in Python. Create a class `ElectricCar` that inherits from both `Car` and `Battery` classes. The `Battery` class should have an attribute for battery capacity and a method to display it. The `ElectricCar` class should have a method to display all the information about the car, including its battery capacity.

class Battery:
    def __init__(self, battery_capacity):
        self.battery_capacity = battery_capacity

    def display_battery_info(self):
        return f"Battery Capacity: {self.battery_capacity} kWh"

class Car:
    def __init__(self, model, doors, windows, max_speed):
        self.model = model
        self.doors = doors
        self.windows = windows
        self.max_speed = max_speed
    
    def get_info(self):
        return f"Model: {self.model}, Doors: {self.doors}, Windows: {self.windows}, Max Speed: {self.max_speed} km/h"


class ElectricCar(Car, Battery):
    def __init__(self, model, doors, windows, max_speed, battery_capacity):
        Car.__init__(self, model, doors, windows, max_speed)
        Battery.__init__(self, battery_capacity)

    def display_info(self):
        car_info = self.get_info()
        battery_info = self.display_battery_info()
        return f"{car_info}, {battery_info}"  
    
```

```python
electric_car = ElectricCar("Tesla Model 3", 4, 4, 225, 75)
print(electric_car.display_info())
```

## Polymorphism

Polymorphism allows objects of different types to be treated as instances of the same type through a common interface. It enables a single interface to be used to operate on objects of different types, providing flexibility and extensibility in object-oriented programming.

Method Overriding: When a child class defines a method with the same name and signature as a method in its parent class, it is said to be overriding that method. This allows the child class to provide its own implementation of the method while still maintaining the same interface.

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
    

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        import math
        return math.pi * (self.radius ** 2)
    
rect = Rectangle(5, 10)
circle = Circle(7)
print(f"Area of Rectangle: {rect.area()}")
print(f"Area of Circle: {circle.area()}")
```

### Abstract Classes
Think of an abstract class like a blueprint or template for building something. It's not a complete, usable thing on its own—you can't directly create an object from it. Instead, it defines the basic structure and rules that other classes (called subclasses) must follow. For example, a "Vehicle" abstract class might say all vehicles need a "start" method, but it doesn't explain how to start. Subclasses like "Car" or "Bike" fill in the details.

### Abstract Methods
An abstract method is like a placeholder or a promise in an abstract class. It's declared (named and described) but doesn't have any actual code inside it. The idea is that any subclass must provide the real implementation. Using the vehicle example, the abstract method "start()" forces "Car" to explain how a car starts (turn the key) and "Bike" to explain how a bike starts (pedal or kick). This ensures consistency while allowing flexibility.

```python
from abc import ABC, abstractmethod

class Food(ABC):
    def __init__(self, name, calories):
        self.name = name
        self.calories = calories

    @abstractmethod
    def get_info(self):
        pass

class Pizza(Food):
    def __init__(self, name, calories, toppings):
        super().__init__(name, calories)
        self.toppings = toppings

    def get_info(self):
        return f"Food: {self.name}, Calories: {self.calories}, Toppings: {', '.join(self.toppings)}"

class Salad(Food):
    def __init__(self, name, calories, ingredients):
        super().__init__(name, calories)
        self.ingredients = ingredients

    def get_info(self):
        return f"Food: {self.name}, Calories: {self.calories}, Ingredients: {', '.join(self.ingredients)}"
    
pizza = Pizza("Pepperoni Pizza", 300, ["Pepperoni", "Cheese", "Tomato Sauce"])
salad = Salad("Caesar Salad", 200, ["Lettuce", "Croutons", "Parmesan Cheese"])

print(pizza.get_info())
print(salad.get_info())
```

## Ecapsulation and abstraction

Encapsulation is the concept of bundling data and methods that operate on that data within a single unit, such as a class. It restricts direct access to some of an object's components, which can prevent the accidental modification of data. In Python, we can achieve encapsulation using private attributes and methods by prefixing them with double underscores (__).m

```python
class ATM:
    def __init__(self, bank_name):
        self.name = bank_name
        self.__balance = 0
        self.account_number = None

    def __get_balance(self):
        return self.__balance
    
    def __set_balance(self, amount):
        if amount < 0:
            raise ValueError("Balance cannot be negative.")
        else:
            self.__balance = amount

    def create_account(self, account_number, initial_deposit):
        self.account_number = account_number
        self.__set_balance(initial_deposit)
        return f"Account {self.account_number} created with balance {self.__get_balance()}."
    
    def get_balance(self):
        return f"Account {self.account_number} balance: {self.__get_balance()}"
    
    def deposit(self, amount):
        self.__set_balance(self.__get_balance() + amount)
        return f"{amount} deposited. New balance: {self.__get_balance()}"
    
    def withdraw(self, amount):
        if amount > self.__get_balance():
            return f"Insufficient funds. Your available balance is {self.__get_balance()}."
        else:
            self.__set_balance(self.__get_balance() - amount)
            return f"{amount} withdrawn. New balance: {self.__get_balance()}"
        
atm = ATM("State Bank of India")
print(atm.create_account("123456789", 1000))
print(atm.get_balance())
print(atm.deposit(500))
print(atm.withdraw(10000))    

```

## Operator overloading

Operator overloading allows us to define custom behavior for operators when they are used with instances of a class. This is done by implementing special methods in the class. For example, we can overload the `+` operator to add two objects of a class together.

| Operator | Magic Method  |
| -------- | ------------- |
| `+`      | `__add__`     |
| `-`      | `__sub__`     |
| `*`      | `__mul__`     |
| `/`      | `__truediv__` |
| `==`     | `__eq__`      |
| `>`      | `__gt__`      |


```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"
    

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)   # Vector(6, 8)
print(v1 - v2)   # Vector(-2, -2)
print(v1 * 3)    # Vector(6, 9)
print(v2 * 3)    # Vector(6, 9)
print(v1 == v2)  # False
```
