```python
import os
from dotenv import load_dotenv
load_dotenv()

from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
store = {}

def get_session_id(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

prompt = ChatPromptTemplate([
    ("system",
     "You are an expert tracker and an helpful assistant who keeps track " 
     "all of the books I'm reading and all the movies I completed watching." 
    ),
    MessagesPlaceholder(variable_name = "messages")
])

model = ChatGroq(model="llama-3.1-8b-instant", api_key=groq_api_key)

chain = prompt | model
```

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_id,
    input_messages_key="messages"
)

config= {
    "configurable":{
        "session_id": "chat_1"
    }
}

with_message_history.invoke(
    {
        "messages": HumanMessage("I just finished reading The Great Gatsby")
    },
    config= config
)
```

```python
with_message_history.invoke(
    {
        "messages": HumanMessage("What books have I read?")
    },
    config= config
)
```

```python
with_message_history.invoke(
    {
        "messages": HumanMessage("I watched Inception yesterday")
    },
    config= config
)
```

```python
with_message_history.invoke(
    {
        "messages": HumanMessage("List all the movies and books I've consumed")
    },
    config= config
)
```
