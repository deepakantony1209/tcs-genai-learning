## Embedding Techniques

Embedding techniques are used to convert the loaded data into vector format for faster retrieval and processing.


```python
## Importing keys: This block uses python_dotenv to load environment variables from a .env file, which is commonly used to store sensitive information like API keys. The load_dotenv() function reads the .env file and makes the variables available in the environment, allowing the code to access them securely without hardcoding them in the script.

import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="embeddinggemma:latest")

r1 = embeddings.embed_documents(
    [
    "The cat sits outside",
    "A man is playing guitar",
    "The new movie is awesome"
     ]
    )
```
