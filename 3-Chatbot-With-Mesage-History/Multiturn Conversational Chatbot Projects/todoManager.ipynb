## Personal To-Do List Chatbot with Message History

In this notebook, we will create a personal to-do list chatbot that can remember the user's tasks and provide updates based on the conversation history. We will use Langchain's `RunnableWithMessageHistory` to manage the conversation context effectively.

Input: "Add buy groceries to my todo list"
Follow-up: "What tasks do I have?"
Follow-up: "Mark buy groceries as done"
Follow-up: "Show my remaining tasks"

```python
## Importing all necessary packages and keys

import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")

from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.messages import HumanMessage
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
```

```python
store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are an experienced to do list manager whom I report the tasks to. "
            "You are well versed in extracting the task from the text I give and remember "
            "all the task with relevant info. You are also capable of updating, deleting, "
            "modifying and displaying the task list. You are well versed in identifying the "
            "priority of the task based on the tasks the user gives. Make sure to perform all "
            "these tasks without any hallucinations or errors. When the user asks to display "
            "the tasks you should present it to them as bullets arranged priority wise."
        ),
        MessagesPlaceholder(variable_name="messages"),
    ],
)

llm = ChatGroq(model="llama-3.1-8b-instant", api_key=groq_api_key)

chain = prompt | llm
```

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages"
)

config = {
    "configurable": {
        "session_id": "chat_1"
    }
}

with_message_history.invoke(
    {
        "messages" : HumanMessage("Add buy groceries to my todo list")
    },
    config = config
)
```

```python
with_message_history.invoke(
    {
        "messages" : HumanMessage("Add Turn of the cooker in 5 minutes to my todo list")
    },
    config = config
)
```

```python
with_message_history.invoke(
    {
        "messages" : HumanMessage("Mark Add buy groceries as done")
    },
    config = config
)
```

```python
with_message_history.invoke(
    {
        "messages" : HumanMessage("Show my remaining tasks")
    },
    config = config
)
```
