```python
import os
from dotenv import load_dotenv
load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq
```

```python
store = {}

def get_session(session_id :str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
prompt = ChatPromptTemplate(
    [
        ("system", "You aren an helpful assistant who is well versed in managing accounts and expenses."),
        MessagesPlaceholder(variable_name="messages")
    ]
)

model = ChatGroq(model="qwen/qwen3-32b", api_key= groq_api_key)

parser = StrOutputParser()

chain = prompt | model | parser
```

```python
config = {
    "configurable" : {
        "session_id": "chat1"
    } 
}

response = RunnableWithMessageHistory(
    chain,
    get_session,
    input_messages_key= "messages"
)
```

```python
response.invoke(
    {"messages": [HumanMessage("I spent $50 on groceries today.")]},
    config = config
)
```
