## Lambda functions

Lambda functions are anonymous functions that can be defined in a single line of code. They are often used for short, simple functions that are not reused elsewhere in the code. The syntax for a lambda function is as follows:

There can be any number of arguments, but only one expression. The expression is evaluated and returned when the lambda function is called.

```python
lambda arguments: expression
```

````python
## Conventional functions

def add(x, y):
    return x+y

print(add(2, 3))
````

````python
# Lambda functions for adding two values

sum = lambda x, y: x+y
print(sum(2, 3))
````

````python
## Finding even numbers

def is_even(n):
    if n%2 == 0:
        return True

is_even(4)
````

````python
## Lambda functions

is_even = lambda n: n%2 == 0
print(is_even(4))
````

````python
## Adding three numbers

def add(x, y, z):
    return x+y+z

print(add(2, 3, 4))
````

````python
## Lambda functions

addition = lambda x, y, z: x + y + z
print(addition(2, 3, 4))
````

````python
# Converting Farenheit to Celsius

ftoc = lambda f: (f - 32) * 5.0/9.0
print(ftoc(100))
````

````python
# Check if a number is positive or negative

pos_or_neg = lambda p: "Positive" if p > 0 else ("Negative" if p < 0 else "Zero")
pos_or_neg(-2)
````

````python
# Calculates the area of rectangle

area = lambda l, b: 2 * (l + b)
print(area(12, 8))
````

````python
# Calculate the number of vowels
# del count_vowels
count_vowels = lambda s: sum(1 for char in s if char.lower() in 'aeiou')
print(count_vowels("vfdf"))
````

## Map functions: 
The `map()` function applies a given function to each item of an iterable (like a list) and returns a map object (which is an iterator). The syntax for the `map()` function is as follows:

```python
map(function, iterable)
```

````python
def square(x):
    return x*x

list(map(square, [1,2,3,4,5,6]))
````

````python
## Lambda functions with map

numbers = [1, 2, 3, 4, 5]
list(map(lambda x: x*2, numbers))
````

````python
array1 = [1,2,3]
array2 = [4,5,6]

list(map(lambda x,y: x+y, array1, array2))
````

````python
## Map function to convert list of strings to integer

array = ['1', '2', '3', '4']

list(map(lambda x: int(x), array))

# You can also use the inbuilt int() functions

list(map(int, array))
````

````python
words = ['apple', 'banana', 'cherry']
upper_word = list(map(str.upper, words))
upper_word
````

````python
people = [
    {
        'name': 'Deepak',
        'age' : 26,
        'spouse': 'Lovely',
     },
     {
         'name': 'Lovely',
         'age': 26,
         'spouse': 'Deepak',
     },
]

def get_name(person):
    return person['name']

result = list(map(get_name, people))

print(f"{result[0]} weds {result[1]}")
````

## Filter

The `filter()` function constructs an iterator from elements of an iterable for which a function returns true. The syntax for the `filter()` function is as follows:

```python
filter(function, iterable)
```

````python
def is_even(n):
    if n%2 == 0:
        return True
    
lst = [1,2,3,4,5,6,7,8,9]
````

````python
list(filter(is_even, lst))
````

````python
numbers = [0.1,1,5,4,3,6,8,10,11, 5.13]
lst = list(filter(lambda x:x>5, numbers))
sort = sorted(lst)
sort
````

## Conclusion

Lambda functions, along with the `map()` and `filter()` functions, provide powerful tools for functional programming in Python. They allow for concise and efficient code when working with iterables and can help improve readability when used appropriately.

Lambda functions are particularly useful for creating small, one-off functions that are not needed elsewhere in the code. They can be used in conjunction with `map()` and `filter()` to perform operations on lists and other iterables in a more functional programming style.

Map functions and filter functions are often used in conjunction with lambda functions to perform operations on lists and other iterables in a more functional programming style.

Filter functions are particularly useful for filtering out elements from a list based on certain conditions, while map functions are useful for applying a function to each element in a list to transform it in some way.
