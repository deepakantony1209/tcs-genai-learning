## Modules and packages in Python

In Python, a module is a file containing Python definitions and statements. A package is a way of organizing related modules into a single directory hierarchy. Modules and packages allow you to organize your code and reuse it across different projects.

```python
# When you use `import math`, you are importing the entire `math` module, and you can access its functions and constants using the module name as a prefix. 
# For example, to calculate the square root of a number using the `sqrt` function from the `math` module, you would write:

import math
math.sqrt(16)
```

```python
# However, if you only need to use specific functions or constants from the `math` module, you can import them directly using the `from` keyword. For example, to import only the `sqrt` function and the constant `pi`, you would write:

from math import sqrt, pi
print(sqrt(14))
print(pi)
```

```python
import numpy as np
print(np.array([1, 2, 3]))
```

## Standard libraries

Python comes with a large standard library that includes modules for various tasks, such as file I/O, regular expressions, and mathematical operations. You can import these modules to use their functionality in your code.

```python
## Difference between array and list
## An array is a data structure that can hold a fixed number of values of the same type, while a list is a more flexible data structure that can hold a variable number of values of different types. Arrays are typically more efficient for numerical operations, while lists are more versatile for general-purpose programming.

import array
print(array.array('i', [1, 2, 3, 4, 5]))
```

```python
## Math module
## The `math` module in Python provides a wide range of mathematical functions and constants. It includes functions for basic arithmetic operations, trigonometric functions, logarithmic functions, and more. You can use the `math` module to perform various mathematical calculations in your Python code.

import math
print(math.sqrt(25))
print(math.sin(math.pi / 2))
```

```python
## OS module
## The `os` module in Python provides a way of using operating system dependent functionality. It allows you to interact with the file system, manage processes, and perform other tasks related to the operating system. You can use the `os` module to create directories, list files, and execute system commands, among other things.

import os
print(os.getcwd())
```

```python
## Shutil module
## The `shutil` module in Python provides a higher-level interface for file operations. It allows you to copy, move, and delete files and directories, as well as perform other file-related tasks. You can use the `shutil` module to easily manage files and directories in your Python code.

import shutil
shutil.copy('data\\source_file.txt', 'data\\destination_file.txt')
```

```python
## Data Serialization
## The `json` module in Python provides functions for working with JSON (JavaScript Object Notation) data. It allows you to convert Python objects to JSON format and vice versa. You can use the `json` module to serialize Python objects into JSON strings and deserialize JSON strings back into Python objects.

import json
data = {'name': 'Alice', 'age': 30, 'city': 'New York'}
json_str = json.dumps(data)  # Serialize Python object to JSON string
print(json_str)

data_dict = json.loads(json_str)  # Deserialize JSON string back to Python object
print(data_dict)
```

```python
## CSV module
## The `csv` module in Python provides functionality for reading and writing CSV (Comma Separated Values) files. It allows you to easily work with tabular data stored in CSV format. You can use the `csv` module to read data from CSV files into Python objects and write Python objects back to CSV files.

import csv
with open('data\\example.csv', mode='w', newline='') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['Name', 'Age', 'City'])
    csv_writer.writerow(['Alice', 30, 'New York'])
with open('data\\example.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)
```

```python
## Datetime module
## The `datetime` module in Python provides classes for manipulating dates and times. It allows you to work with date and time objects, perform arithmetic operations on them, and format them in various ways. You can use the `datetime` module to handle date and time data in your Python code.

import datetime
current_datetime = datetime.datetime.now()
print(current_datetime)

yesterday = current_datetime - datetime.timedelta(days=1)
print(yesterday)
```

```python
## time module
## The `time` module in Python provides functions for working with time-related tasks. It allows you to measure time intervals, pause the execution of your program, and perform other time-related operations. You can use the `time` module to get the current time, sleep for a specified duration, and more.

import time
print(time.time())
time.sleep(2)  # Sleep for 2 seconds
print("This message is printed after a 2-second delay.")
print(time.time())
```

```python
## Regex module
## The `re` module in Python provides functions for working with regular expressions. Regular expressions are a powerful tool for searching and manipulating strings based on specific patterns. You can use the `re` module to perform tasks such as searching for substrings, replacing text, and validating input against a pattern.

import re

pattern = r'\d+'  # Pattern to match one or more digits
text = "The price is 100 dollars."
matches = re.findall(pattern, text)
print(matches)  # Output: ['100']
```
