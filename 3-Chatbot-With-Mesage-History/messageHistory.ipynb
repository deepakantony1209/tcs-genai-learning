 # Gen AI functionality with Message History

 In this notebook, we will see how to use the message history functionality of Langchain to keep track of the conversation history. We will use the Groq API as our LLM and we will use the StrOutputParser to parse the output of the LLM. We will also see how to use the prompt template to format the input and output of the LLM. 

```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_core.output_parsers import StrOutputParser
load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
model = ChatGroq(model="llama-3.1-8b-instant", api_key=groq_api_key)
model.invoke([
    SystemMessage(content="You are my personal assistant who is very helpful and always answers in a very concise manner."),
    HumanMessage(content="Im Deepak, Im a Generative AI engineer. My aadhar id is 907714234750"),
])
```

```python
model.invoke([
    HumanMessage(content="Im Deepak, Im a Generative AI engineer. My aadhar id is 907714234750"),
    AIMessage(content="Nice to work with you, Deepak. Your Aadhar ID is 9077 1423 4750. I'll keep this secure and confidential. How can I assist you  today with your AI engineering work?"),
    HumanMessage(content="Who am I"),
])
```

## Message History

We can use message history class to wrap our model and make it stateful. This will allow us to keep track of inputs and outputs of the model and store them in the datastore. Future interactions will then load those messages and pass them into the chain as a part of the input.  

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model, get_session_history)
config = {
    "configurable": {
        "session_id": "chat1"
        },
}
```

```python
with_message_history.invoke([
    HumanMessage(content="Im Deepak, Im a Generative AI engineer"),
], config=config)
```

```python
with_message_history.invoke([
    HumanMessage(content="Who am I, what is are my possible skills"),
], config=config)
```

```python
config1 = {
    "configurable": {
        "session_id": "chat2"
        },
}
```

```python
with_message_history.invoke([
    HumanMessage(content="Who am I, what is are my possible skills"),
], config=config1)
```
