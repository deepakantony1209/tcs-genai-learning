## Chroma
Chroma is a popular vector database that is designed to be easy to use and integrate with various applications. It provides a simple API for storing and querying vector data, making it a great choice for applications that require fast and efficient similarity search.

```python
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
## Data Ingestion

loader = TextLoader("data.txt")
data = loader.load()

## Data Transformation
text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=30)
docs = text_splitter.split_documents(data)

## Embeddings and Vector Store
embeddings = OllamaEmbeddings(model="embeddinggemma:latest")
db = Chroma.from_documents(docs, embeddings, persist_directory="./chroma_db")
```

```python
## Similarity Search in Vector Store

db = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
results = db.similarity_search_with_score("What is the assurance given by Tamizhaga Vetri Kazhagam")
results
```

```python
## Retrivers

retriever = db.as_retriever()
query = "What is the assurance given by Tamizhaga Vetri Kazhagam"

docs = retriever.invoke(query)

```
