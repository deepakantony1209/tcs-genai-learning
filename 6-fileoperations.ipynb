## File operations in Python

In Python, you can perform various file operations such as reading, writing, and appending to files. You can use the built-in `open()` function to work with files. The `open()` function takes the file name and mode as arguments. The mode can be 'r' for reading, 'w' for writing (which will overwrite the file), and 'a' for appending (which will add to the end of the file). After performing the desired operations, it's important to close the file using the `close()` method or by using a context manager (the `with` statement) to ensure that resources are properly released.

```python
# Example of writing to a file
with open('example.txt', 'w') as file:
    file.write('Hello, World!\n')
    file.write('This is a file operation example.')

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```
