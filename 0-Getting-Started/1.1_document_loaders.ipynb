## Text splitter

Text splitter is used to split the documents into smaller chunks. This is useful when we want to use the documents as input to a language model, which has a maximum token limit. The text splitter will split the documents into chunks of a specified size, with a specified overlap between the chunks. This way, we can ensure that we don't lose any important information when splitting the documents.

```python
from langchain_community.document_loaders import PyPDFLoader
pdf_loader = PyPDFLoader("attention.pdf").load()
pdf_loader
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=70)
pdf_chunks = text_splitter.split_documents(pdf_loader)
pdf_chunks
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader = TextLoader("data.txt").load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=50)
text_chunks = text_splitter.split_documents(loader)
text_chunks
```

## Character Splitter

Character splitter is a simple text splitter that splits the documents based on a specified character. This is useful when we want to split the documents based on a specific character, such as a newline or a period. The character splitter will split the documents into chunks based on the specified character, with a specified overlap between the chunks. This way, we can ensure that we don't lose any important information when splitting the documents. 

```python
from langchain_text_splitters import CharacterTextSplitter
character_splitter = CharacterTextSplitter(separator="\n\n", chunk_size=100, chunk_overlap=50)
character_chunks = character_splitter.split_documents(loader)
```

```python
print(character_chunks[0])
print(character_chunks[1])
```

## Html Header Text Splitter

HTML header text splitter is a text splitter that splits the documents based on the HTML headers. This is useful when we want to split the documents based on the structure of the HTML document. The HTML header text splitter will split the documents into chunks based on the HTML headers, with a specified overlap between the chunks. This way, we can ensure that we don't lose any important information when splitting the documents.

```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_content = """
Generate a complex HTML content
<!DOCTYPE html>
<html>
<head>
    <title>Sample HTML Document</title>
</head>
<body>
    <h1>Welcome to the Sample HTML Document</h1>
    <p>This is a sample paragraph to demonstrate the HTML header text splitter.</p>
    <h2>Section 1</h2>
    <p>This is the first section of the document.</p>
    <h2>Section 2</h2>
    <p>This is the second section of the document.</p>
</body>
</html>
"""

headers_to_split = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split)
html_chunks = html_splitter.split_text(html_content)
html_chunks
```

```python
import json
import requests

json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter = RecursiveJsonSplitter(max_chunk_size = 300)
json_chunks = json_splitter.split_json(json_data)

for chunk in json_chunks[:3]:
    print(chunk)
```

```python
docs = json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(doc)
```

```python
text = json_splitter.split_json(json_data)
print(text[0])
print(text[1])
print(text[2])
```
