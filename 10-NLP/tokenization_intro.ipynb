## Stemming

Stemming in NLP is a text preprocessing technique that reduces words to their base or root form (e.g., "running" to "run") by crudely removing suffixes and prefixes. It standardizes vocabulary, reduces data dimensionality, and improves efficiency in information retrieval and text classification, although it may produce non-dictionary root words.

```python
## Let's say you have to classify a reviews
## You have to know whether the review is positive or negative
## You can use stemmer to reduce the words to their root form
## For example, "running", "runs", "ran" can be reduced to "run

words = ["running", "runs", "ran", "jumping", "jumps", "jumped", "swimming", "swims", "swam", "eating", "happily"]
words
```

```python
## PorterStemmer
"""
It is a popular stemming algorithm that reduces words to their root form by applying a series of rules. 
For example, it will reduce "running", "runs", and "ran" to "run". 
This helps in text analysis and natural language processing tasks by treating different forms of a word as the same base word.

Pros:
1. It is a widely used and well-known stemming algorithm.
2. It is effective in reducing words to their root form, which can improve the performance of text analysis tasks.
3. It is relatively fast and efficient.

Cons:
1. It can sometimes produce non-dictionary words as stems, which may not be ideal for certain applications.
2. It may not always produce the correct root form, especially for irregular verbs or words with multiple meanings.
3. It may not be suitable for all languages, as it is primarily designed for English and may not work well with other languages that have different morphological structures.
"""


from nltk.stem import PorterStemmer
ps = PorterStemmer()
for word in words:
    print(ps.stem(word))
```

```python
## RegexStemmer
## Write about regex stemmer and its pros and cons
"""
The RegexStemmer is a simple stemming algorithm that uses regular expressions to identify and remove common suffixes from words.
For example, it can be configured to remove suffixes like "ing", "s", and "ed" from words, which can help in reducing them to their root form.
Pros:
1. It is easy to implement and customize based on specific requirements.
2. It can be effective for certain types of text where specific suffixes are common.
3. It can be faster than more complex stemming algorithms for certain applications.

Cons:
1. It may not be as effective as more sophisticated stemming algorithms, as it relies on simple pattern matching and may not account for irregular forms or exceptions.
2. It may produce incorrect stems if the regular expressions are not carefully designed, leading to loss of meaning or incorrect analysis.
3. It may not be suitable for all languages or types of text, as it may not account for the complexities of different morphological structures or linguistic rules. 
"""


from nltk.stem import RegexpStemmer
rs = RegexpStemmer('ing$|s$|ed$')
for word in words:
    print(rs.stem(word))
```

```python
## SnowballStemmer
## Write about SnowballStemmer and its pros and cons
"""

The SnowballStemmer is an improved version of the Porter Stemmer that provides better stemming results for a wider range of languages.
Pros:
1. It is an improved version of the Porter Stemmer, providing better stemming results.
2. It supports multiple languages, making it more versatile for different applications.
3. It is designed to be more efficient and accurate than the Porter Stemmer.

Cons:
1. It may still produce non-dictionary words as stems, which may not be ideal for certain applications.
2. It may not always produce the correct root form, especially for irregular verbs or words with multiple meanings.
3. It may be more complex to implement and understand compared to simpler stemming algorithms.

"""

from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
for word in words:
    print(ss.stem(word))
```
